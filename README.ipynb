# Material Symbols SVG → PNG (Brand Icon Export)

This notebook converts **Google Material Symbols SVG icons** into **PNG files** in multiple **brand colors** and **sizes**, optimized for dashboards (e.g. **Power BI**), presentations, and reports.

---

## What this notebook does
- Reads SVG icons from a **folder or ZIP file**
- Recolors icons into a **custom brand color palette**
- Exports crisp PNGs at one or more sizes (recommended: **48px**)
- Organizes output into **per-color folders** (and optional size subfolders)
- Uses **CairoSVG** for reliable rendering of Material Symbols

## Requirements
- Python 3.9+
- Packages:
  - `cairosvg`
  - `pillow`
  - `tqdm`

> **Windows users:** installing via **conda-forge** is recommended for best compatibility.

## How to use
1. Place all `.svg` icons in a folder **or** prepare a ZIP containing SVG files.
2. Open this notebook and edit the **Configuration** cell:
   - Set the input path (folder or ZIP)
   - Set the output directory
   - Define brand colors and icon sizes
3. Run all cells.
4. PNG icons will be written to the output directory.

## Recommended settings (Power BI)
- Icon size: **48px**
- Background: **transparent**
- Folder layout: `COLOR/size_48/icon.png`

## Customization
- **Add colors:** append hex codes to the `COLORS` list
- **Change sizes:** edit the `SIZES` list (e.g. `[24, 48, 96]`)
- **Folder structure:** toggle `NEST_BY_SIZE`

## Example output structure
```
IconsPNG_Export/
├── F05A28/
│   └── size_48/
│       └── analytics.png
├── 204050/
│   └── size_48/
│       └── analytics.png
```

## Notes
- Power BI scales images **down** better than it scales up.
- Material Symbols SVGs are best rendered with **CairoSVG**, which this notebook uses.
